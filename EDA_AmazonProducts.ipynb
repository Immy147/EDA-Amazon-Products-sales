# Amazon products sales EDA 2024
In 2024, an exploratory data analysis (EDA) was conducted on Amazon product sales data to uncover trends, patterns, and insights within the dataset. The primary objective was to understand the sales dynamics, identify key drivers of revenue, and analyze consumer behavior.

Key Aspects Covered in the EDA:
Data Cleaning and Preprocessing:

Handled missing values, duplicates, and inconsistencies in the dataset.
Converted relevant columns to appropriate data types, ensuring accuracy in analysis.
Sales Distribution Analysis:

Analyzed the distribution of sales across different product categories, price ranges, and time periods.
Identified top-performing products and categories based on total sales and unit volume.
Time Series Analysis:

Examined monthly, quarterly, and yearly sales trends to identify seasonal patterns and growth rates.
Detected sales spikes during significant sales events like Black Friday, Cyber Monday, and Prime Day.
Customer Behavior Insights:

Analyzed customer reviews, ratings, and feedback to assess product satisfaction and potential areas for improvement.
Investigated the relationship between product pricing, ratings, and sales performance.
Geographical Sales Distribution:

Mapped sales data to various regions and countries to determine market strengths and areas with growth potential.
Explored correlations between regional preferences and product categories.
Competitive Analysis:

Compared sales performance with competitors to identify market position and competitive advantages.
Analyzed pricing strategies, discounts, and promotions impacting sales.

```python
df.isnull().sum()
```

```python
perc = 75.0
min_count =  int(((100 - perc)/100) * df.shape[1] + 1)
mod_df = df.dropna(axis = 1, thresh = min_count)
```

```python
mod_df.columns
```

```python
mod_df.info()
```

```python
mod_df.describe()
```

```python
mod_df["discount_price"] = mod_df["discounted_price"].str.split(" ", expand = True).get(0).str.split("₹", expand = True).get(1)
mod_df["actual_price"] = mod_df["actual_price"].str.split(" ", expand = True).get(0).str.split("₹", expand = True).get(1)
```

```python
mod_df['discount_price'] = mod_df["discount_price"].str.replace(',', '').astype(float)
mod_df["actual_price"] = mod_df["actual_price"].str.replace(',', '').astype(float)
```

```python
mod_df["rating"].unique()
```

```python
mod_df['ratings'] = mod_df['rating'].replace(['Get','FREE','₹68.99', '₹65','₹70', '₹100', '₹99', '₹2.99'], '0.0')
mod_df['ratings'] = mod_df["rating"].astype(str)
mod_df['ratings'].unique()
```

```python
mod_df['no_of_ratings'] = mod_df['no_of_ratings'].astype(str)
mod_df['correct_no_of_ratings'] = pd.Series([mod_df['no_of_ratings'][x][0].isdigit() for x in range(len(mod_df['no_of_ratings']))])
# Drop columns with incorrect 'no_of_ratings'
mod_df = mod_df[mod_df['correct_no_of_ratings'] == True]
mod_df['correct_no_of_ratings'].value_counts()
```

```python
mod_df["no_of_ratings"] = mod_df["no_of_ratings"].str.replace(',', '').astype(float)
```
